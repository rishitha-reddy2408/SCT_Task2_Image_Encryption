```python
def swap_pixels(image_array):

    swapped = image_array.copy()

    swapped[:,:,0], swapped[:,:,1] = (
        image_array[:,:,1],
        image_array[:,:,0]
    )

    return swapped


swapped_array = swap_pixels(img_array)

swapped_image = Image.fromarray(
    swapped_array.astype('uint8')
)


plt.figure(figsize=(5,5))
plt.imshow(swapped_image)
plt.axis("off")
plt.title("Pixel Swapped Image")
plt.show()
```
